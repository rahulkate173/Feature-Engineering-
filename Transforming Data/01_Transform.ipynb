### NORMAL/GUASSIAN TRANSFORMATION

### 1. Logarthimic Transformation

skewed distribution is converted to normal distribution

```python
import pylab
import scipy.stats as stat
```

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],bins=20,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Fare')
```

--> Not good because not much more or data points are on line(red)

```python
plot_data(df,'Age')
```

--> It is good

```python
df['Log_Fare']=np.log1p(df['Fare'])
plot_data(df,'Log_Fare')
```

--> Used log1p() Beacause it is containing zero(0) value log(0)=infinity

```python
df['Age_log']=np.log1p(df['Age'])
plot_data(df,'Age_log')
#noramlly skewed data is not left skewed
```

### 2.Reciprocal transformation 

```python
df['Fare_reciprocal']=1/df['Fare']
plot_data(df,'Fare_reciprocal')
```

### 3. Squareroot Transformation

```python
df['Fare_Sqroot']=df['Fare']**(0.5)
plot_data(df,'Fare_Sqroot')
```

```python
df['Age_sqrt']=df['Age']**(0.5)
plot_data(df,'Age_sqrt')
```

--> Okay,that is good

### 4. Exponential Transformation

```python
df['Fare_exp']=df.Fare**(1/1.2)
plot_data(df,'Fare_exp')
```

```python
df['Age_exp']=df.Age**(1/1.2)
plot_data(df,'Age_exp')
```

### 5. Box Cox Transformation

    T(y)=(y.exp(l)-1)/l

    - l:lambda(parameter)
    - y:response variable or what is given

```python
df['Age_boxcox'],parameter=stat.boxcox(df['Age'])
plot_data(df,'Age_boxcox')
```

```python
df.Fare[df.Fare==0]=1
df['Fare_boxcox'],parameter=stat.boxcox(df['Fare'])
plot_data(df,'Fare_boxcox')
```
