```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(pd.date_range(start='01/01/2025',periods=30,freq='D'))
```

```python
df.columns=['Day']
```

```python
df
```

```python
import datetime
df['Day']=pd.to_datetime(df['Day'])
df['Weekday']=df['Day'].dt.day_name()
```

```python
weekday_map = {'Monday':1,
               'Tuesday':2,
               'Wednesday':3,
               'Thursday':4,
               'Friday':5,
               'Saturday':6,
               'Sunday':7
}
df['ordinal Encoding']=df['Weekday'].map(weekday_map)
```
