### 2. One Hot Encoding For Multivariable Category

```python
df=pd.read_csv('D:/OneDrive/Desktop/jupyter/Feature Enginerring/datasets/mercedesbenz.csv',usecols=['X0','X1','X2','X3','X4','X5'])
```

```python
df.head()
```

```python
listX0_10=list(df.X0.value_counts().sort_values(ascending=False).head(10).index)
listX0_10
```

```python
def MultivariableEncoding(df,variable):
    for label in listX0_10:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
MultivariableEncoding(df,'X0')
```

### 6. Probability Ratio Encoding 

```python
df=pd.read_csv('D:/OneDrive/Desktop/jupyter/Feature Enginerring/Datasets/titanic.csv',usecols=['Survived','Cabin'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin'].unique()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()
prob_df
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Not Survived']=1-prob_df['Survived']
```

```python
prob_df['Porbability Encoding']=prob_df['Survived']/prob_df['Not Survived']
```

```python
prob_df.head()
```

```python
encoded_dict=prob_df['Porbability Encoding'].to_dict()
```

```python
df['Probability Encoding']=df['Cabin'].map(encoded_dict)
```
