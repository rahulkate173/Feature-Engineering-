```python
import pandas as pd
import numpy as np
df=pd.read_csv("mercedesbenz/mercedesbenz.csv",usecols=['X0','X1','X2','X3','X4','X5','X6'])
df.head()
```

```python
pd.get_dummies(df,drop_first=True).shape
```

```python
data=df[['X0','X1','X2','X3','X4','X5','X6']]
```

```python
for col in data.columns:
    print(col,':',len(data[col].unique()),'labels')
    
```

```python
pd.get_dummies(data,drop_first=True).shape
```

```python
data['X2'].value_counts().sort_values(ascending=False).head(10)
```

```python
top10X2=[x for x in data['X2'].value_counts().sort_values(ascending=False).head(10).index]
top10X2
```

```python
for label in top10X2:
    data[label]=np.where(data['X2']==label,1,0)
data[['X2']+top10X2]
```

```python
def one_hot_encoding(df,variable,top_x_label):
    for label in top_x_label:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
    

one_hot_encoding(data,'X2',top10X2)
```
